Pelo enunciado, sabemos que a memória principal tem 4K de blocos com 128 bytes em cada um. Assim, se multiplicarmos um pelo outro, teremos a quantidade total de bytes na memória principal.

```python
import math

Ncelulas = 4*2**10*2**7

print(str(Ncelulas) + " células.")
```

Tendo esse número, podemos calcular a quantidade de bits necessária para endereçar cada byte:

```python
bitsEnd = math.log2(Ncelulas)

print(str(bitsEnd) + " bits.")
```

Ou seja, teremos um barramento de endereço da MP igual a 19 bits.

Para a memória cache:

```python
linhas = 256
qtdConjunto = linhas/2
bitsConjunto = math.log2(qtdConjunto)
print(bitsConjunto, " bits para conjunto.")
```

```python
qtdByte = 128

bitsByte = math.log2(qtdByte)

print(bitsByte, " bits para byte.")
```

Já temos 7 bits para endereçar o conjunto e 7 linhas para endereçar o byte (total de 14 bits). Como o barramento de endereços da MP tem o total de 19 bits, temos que os 5 bits restantes são para endereçar a tag.

Quando o número de conjuntos de uma memória associativa por conjuntos for igual ao número de linhas dessa mesma memória, na prática ela se transforma em uma memória por mapeamento direto.

```python

nCelulas = 2**6*2**10
celulasBloco = 2**3

nBlocos = nCelulas/celulasBloco

print(nBlocos, " blocos.")
```

```python
bitsBloco = math.log2(nBlocos)
bitsMP = math.log2(nCelulas)

print("Para endereçar a MP temos ", bitsMP, " bits, e para endereçar o bloco temos ",
      bitsBloco, " bits. Assim, sabemos que temos" ,bitsMP - bitsBloco, " bits para endereçar o byte.")
```

Também sabemos que há 32 linhas, assim:

```python
bitsLinha = math.log2(32)

print("Temos ", bitsLinha, " bits para endereçar a linha. Dessa forma, sabemos que temos", bitsBloco - bitsLinha, 
      " bits para endereçar a tag.")
```

```python
linhas = 2**5
byt = 2**3

capacidadeTotal = byt * linhas

print("Capacidade total da memória cache: ", capacidadeTotal, " bytes.")
```

Isso se deve pela maneira em que a memória cache é organizada com relação aos blocos e ao sistema de mapeamento.

O método direto não precisa desse sistema pelo fato de que cada bloco já pertence a uma linha especifica. Assim, quando há um outro bloco que pertence a essa mesma linha, a memória vai apenas substituir no devido lugar. Os métodos associativos precisam determinar qual linha vai ser substituída de alguma maneira pois não há linha específica para os blocos.

```python
capacidadeTotalMP = 4*2**10*128
bitsEndMP = math.log2(capacidadeTotalMP)

bitsConj = math.log2(64/4)

bitsByte = math.log2(128)

print("Temos", bitsEndMP, " bits para endereçar a MP. Destes, como temos ", bitsByte, " bits para endereçar os bytes e ",
     bitsConj, " bits para endereçar o conjunto, temos os ", bitsEndMP - (bitsConj + bitsByte),
      " bits restantes para endereçar a tag.")
```

Como o endereço especificado, em binário, é: 0101 1101 0111 1010 1001 1111 0010. Assim, a linha especificada é a linha 111 1010 1001 111.

```python
bitsCapacidadeTotalMP = math.log2(64*2**20)
bitsByte = math.log2(32)
nLinhasCache = 64*2**10/32
bitsConjunto = math.log2(nLinhasCache/4)

print("Temos ", bitsByte, " para endereçar o byte e ", bitsConjunto, " bits para endereçar o conjunto. Assim, temos ",
     bitsCapacidadeTotalMP - (bitsByte + bitsConjunto), " bits para endereçar a tag.")
```

Mapeamento direto:

```python
CapacidadeTotalMP = 2**30
CapacidadeDadosMC = 256*2**10*8
nBlocos = CapacidadeTotalMP/16
nLinhas = 256*2**10/16
bitsTag = math.log2(nBlocos/nLinhas)

CapacidadeTotalMC = CapacidadeDadosMC + bitsTag * nLinhas

print("O total de bits da memória cache para o mapeamento direto é: ", CapacidadeTotalMC, " bits.")
```

Mapeamento Associativo:

```python
CapacidadeTotalMP = 2**30
CapacidadeDadosMC = 256*2**10*8
nBlocos = CapacidadeTotalMP/16
nLinhas = 256*2**10/16
bitsTag = math.log2(nBlocos)

CapacidadeTotalMC = CapacidadeDadosMC + bitsTag * nLinhas

print("O total de bits da memória cache para o mapeamento direto é: ", CapacidadeTotalMC, " bits.")
```

Mapeamento Associativo por conjunto:

```python
CapacidadeTotalMP = 2**30
CapacidadeDadosMC = 256*2**10*8
nBlocos = CapacidadeTotalMP/16
nLinhas = 256*2**10/16
nConjuntos = 4
bitsTag = math.log2(nBlocos/4)

CapacidadeTotalMC = CapacidadeDadosMC + bitsTag * nLinhas

print("O total de bits da memória cache para o mapeamento direto é: ", CapacidadeTotalMC, " bits.")
```
